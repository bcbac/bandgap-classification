```python
import pandas as pd
import numpy as np
```

```python
from matminer.utils.io import load_dataframe_from_json
df = load_dataframe_from_json('2fABL3.json')
df
```

```python
from matminer.featurizers.structure import StructuralHeterogeneity
strh = StructuralHeterogeneity()
```

```python
df = strh.featurize_dataframe(df, 'structure')
```

```python
df
```

```python
from matminer.featurizers.conversions import StrToComposition
stc = StrToComposition()
df = stc.featurize_dataframe(df, 'pretty_formula')
```

```python
from matminer.featurizers.composition import Meredig
mer = Meredig()
mer.feature_labels()
```

```python
df = mer.featurize_dataframe(df, 'composition')
```

```python
df
```

```python
P = df.drop(['pretty_formula', 'material_id', 'composition', 'structure', 'energy_per_atom', 'target','total_magnetization','formation_energy_per_atom','avg f valence electrons','frac f valence electrons'], axis=1)
P
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
```

```python
Q = df['target']
Q
```

```python
Pt = ss.fit_transform(P)
P = pd.DataFrame(Pt, columns=P.columns)
P
```

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
correlated_features=set()
correlation_matrix=P.corr()
```

```python
correlation_matrix
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

corr = P.corr().round(2)
plt.figure(figsize=(50,50))
#绘制热力图
sns.heatmap(corr,linewidths=0.1,square = True,annot = True)
plt.show()
```

```python
for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i,j]) > 0.8:
            colname = correlation_matrix.columns[i]
            correlated_features.add(colname)
```

```python
print(correlated_features)
```

```python
P = P.drop(labels = correlated_features,axis = 1)
P
```

```python
from sklearn.model_selection import train_test_split, RandomizedSearchCV, cross_val_score
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, roc_curve, auc
```

```python
from xgboost import XGBClassifier
import numpy as np

xgb = XGBClassifier(random_state = 0)
```

```python
params = {'max_depth': [3, 5, 6, 10, 15, 20],
           'learning_rate': [0.01, 0.1, 0.2, 0.3],
           'subsample': np.arange(0.5, 1.0, 0.1),
           'colsample_bytree': np.arange(0.4, 1.0, 0.1),
           'colsample_bylevel': np.arange(0.4, 1.0, 0.1),
           'n_estimators': [100, 500, 1000]}
```

```python
xgbc = RandomizedSearchCV(xgb, params,n_iter=20, cv = 10, verbose=1, scoring='accuracy', random_state=0, n_jobs=-1)
```

```python
xgbc.fit(P,Q)
```

```python
xgbc.best_params_
```

```python
xgb = XGBClassifier(random_state=0, n_estimators=500, max_depth=5, learning_rate=0.01, colsample_bytree=0.7, colsample_bylevel=0.5, subsample=1)
```

```python
xgb
```

```python
arr = cross_val_score(xgb, P,Q, cv=10)
arr
```

```python
np.mean(arr)
```

# 测试集上验证

```python
X_train,X_test,y_train,y_test = train_test_split(P,Q,test_size=0.2, random_state=92)
```

# XGB

```python
xgb
xgb.fit(X_train,y_train)
```

```python
y_xgb = xgb.predict(X_test)
```

```python
confusion_matrix(y_test, y_xgb, labels=[1,0])
```

```python
print(classification_report(y_test, y_xgb))
```

```python
fpr_xgb, tpr_xgb, thr_xgb = roc_curve(y_test, y_xgb)
auc_xgb = auc(fpr_xgb, tpr_xgb)
s1=sns.heatmap(confusion_matrix(y_test, y_xgb, labels=[1,0]),cmap='viridis',fmt = '.2f', annot=True, annot_kws={"size": 16},xticklabels=['Predicted 1','Predicted 0'],yticklabels=['True 1','True 0'])
fig1=s1.get_figure()
fig1.savefig("xgb.svg", dpi=600,format="svg")
```

```python
auc_xgb 
```

```python
fpr_xgb, tpr_xgb, thr_xgb = roc_curve(y_test, y_xgb)
auc_xgb = auc(fpr_xgb, tpr_xgb)
plt.figure()
lw = 2
plt.figure(figsize=(4.5,4))
plt.plot(fpr_xgb, tpr_xgb, color='darkorange',lw=lw,label=' ROC curve (area = %0.2f)' % auc_xgb)
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0,1.0])
plt.ylim([0.0,1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc="lower right")
plt.savefig("1xgb.svg", dpi=600,format="svg")
plt.show()
```

```python
arr = cross_val_score(xgb, y_test,y_xgb, cv=10)
arr
```

```python
np.mean(arr)
```
